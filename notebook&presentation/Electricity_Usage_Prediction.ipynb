````python
import pandas as pd

def get_cities_by_state(cities_df, state_name):
    """
    Filters the cities DataFrame to return only cities in the specified state along with their latitude and longitude.

    Parameters:
        cities_df (pd.DataFrame): DataFrame containing city-level data, including columns for 'state_name', 'lat', and 'lng'.
        state_name (str): The name of the state to filter by.

    Returns:
        pd.DataFrame: A DataFrame of cities in the specified state with their latitude and longitude.
    """
    # Filter cities based on the provided state name
    state_cities_df = cities_df[cities_df['state_name'] == state_name][['city', 'lat', 'lng']]
    
    # Reset the index for clarity
    state_cities_df = state_cities_df.reset_index(drop=True)
    
    return state_cities_df

# Example usage
state_name = 'Massachusetts'  # Replace with the state you're interested in
massachusetts_cities = get_cities_by_state(cities_df, state_name)
print(massachusetts_cities.head())
````

````python
# Fetching Monthly Average Weather Data for a City Function

from datetime import datetime
from meteostat import Point, Daily
import pandas as pd

def fetch_monthly_avg_weather_data(lat, lon, start_date, end_date):
    """
    Fetch monthly average weather data for a given latitude, longitude, and date range.
    
    Parameters:
        lat (float): Latitude of the location
        lon (float): Longitude of the location
        start_date (str): Start date in the format 'YYYY-MM-DD'
        end_date (str): End date in the format 'YYYY-MM-DD'
        
    Returns:
        pandas.DataFrame: DataFrame containing the monthly average weather data with columns 'time', 'tavg', 'prcp', and 'snow'.
    """
    # Convert start and end dates to datetime objects
    start = datetime.strptime(start_date, '%Y-%m-%d')
    end = datetime.strptime(end_date, '%Y-%m-%d')
    
    # Create a Point object for the specified location
    location = Point(lat, lon)
    
    # Fetch daily weather data for the location and date range
    data = Daily(location, start, end)
    daily_data = data.fetch()
    
    # Handle case where no data is available
    if daily_data.empty:
        # print(f"No weather data available for lat: {lat}, lon: {lon} in the given range.")
        return pd.DataFrame()
    
    # Resample to monthly averages and reset the index to include 'time'
    monthly_avg_data = daily_data.resample('ME').mean().reset_index()
    
    # Select only the desired columns: 'time', 'tavg', 'prcp', and 'snow'
    monthly_avg_data = monthly_avg_data[['time', 'tavg', 'prcp', 'snow']]
    
    return monthly_avg_data


# Example check for Boston (latitude and longitude of Boston, MA)
lat = 42.3601
lon = -71.0589
start_date = '2015-01-01'
end_date = '2024-12-31'

# Fetch monthly average weather data
monthly_weather_data = fetch_monthly_avg_weather_data(lat, lon, start_date, end_date)

# Display the first few rows of the monthly average data
print(monthly_weather_data.head())
````

````python
# Section 4: Fetching Monthly Average Weather Data for All Cities in a Specified State

from datetime import datetime
import pandas as pd

def fetch_monthly_avg_weather_for_state(cities_df, state_name, start_date, end_date):
    """
    Fetch monthly average weather data for all cities in a specified state.
    
    Parameters:
        cities_df (pandas.DataFrame): DataFrame containing city information with 'state_name', 'city', 'lat', and 'lng' columns
        state_name (str): Name of the state to fetch weather data for
        start_date (str): Start date in the format 'YYYY-MM-DD'
        end_date (str): End date in the format 'YYYY-MM-DD'
        
    Returns:
        pandas.DataFrame: Combined DataFrame containing monthly average weather data for all cities in the specified state
    """
    # Use the get_cities_by_state function to get all cities in the specified state
    state_cities_df = get_cities_by_state(cities_df, state_name)
    
    all_weather_data = []  # List to store each city's monthly average weather data

    # Iterate over each city in the state's DataFrame
    for index, row in state_cities_df.iterrows():
        city_name = row['city']
        lat = row['lat']
        lon = row['lng']
        
       # print(f"Fetching monthly average weather data for {city_name} (Lat: {lat}, Lon: {lon})")
        
        # Fetch monthly average weather data for the current city
        weather_data = fetch_monthly_avg_weather_data(lat, lon, start_date, end_date)
        
        # Check if any data is returned; if empty, continue to the next city
        if weather_data.empty:
            # print(f"No weather data available for {city_name}")
            continue
        
        # Add the city name to the weather data for identification
        weather_data['city'] = city_name
        
        # Append the city's data to the list
        all_weather_data.append(weather_data)
    
    # Concatenate all city weather data into a single DataFrame
    if all_weather_data:
        combined_weather_df = pd.concat(all_weather_data, axis=0)
    else:
        combined_weather_df = pd.DataFrame()  # Return an empty DataFrame if no data is fetched
    
    return combined_weather_df

# Example usage:

# Define start and end date for the weather data
start_date = '2024-01-01'
end_date = '2024-3-28'

# 'cities_df' is DataFrame containing the cities data
combined_weather_df = fetch_monthly_avg_weather_for_state(cities_df, 'Massachusetts', start_date, end_date)

# Display the first few rows of the fetched data
print(combined_weather_df.head())

````

````python
from datetime import datetime
import pandas as pd

def calculate_statewide_monthly_avg_weather(cities_df, state_name, start_date, end_date):
    """
    Fetch and calculate the monthly average weather data for a specified state across all cities.
    
    Parameters:
        cities_df (pandas.DataFrame): DataFrame containing city information with 'state_name', 'city', 'lat', and 'lng' columns
        state_name (str): Name of the state to fetch weather data for
        start_date (str): Start date in the format 'YYYY-MM-DD'
        end_date (str): End date in the format 'YYYY-MM-DD'
        
    Returns:
        pandas.DataFrame: DataFrame containing the monthly average weather data for the entire state
    """
    # Fetch monthly average weather data for all cities in the specified state
    combined_weather_df = fetch_monthly_avg_weather_for_state(cities_df, state_name, start_date, end_date)
    
    # Check if we have any data
    if combined_weather_df.empty:
        # print(f"No weather data available for the state: {state_name}")
        return pd.DataFrame()

    # Select only numeric columns to avoid errors during aggregation
    numeric_data = combined_weather_df.select_dtypes(include=['float64', 'int64'])
    
    # Group by 'time' and calculate the mean for numeric columns only
    state_monthly_avg = numeric_data.groupby(combined_weather_df['time']).mean().reset_index()
    
    return state_monthly_avg

# Example usage
state_name = 'Massachusetts'
start_date = '2024-01-01'
end_date = '2024-03-30'

massachusetts_weather_avg_df = calculate_statewide_monthly_avg_weather(cities_df, state_name, start_date, end_date)

# Display the first few rows of the result
print(massachusetts_weather_avg_df.head())

````

#### 2.5.1 Time-Based Feature Engineering

In this section, we are creating new time-based features to enrich our dataset and provide more insights into how seasonal variations and peak periods affect electricity usage and cost. Here’s an overview of the columns we are adding and their purpose:

- **`season`**: Categorizes each month into a traditional season:
  - **Winter** (December, January, February)
  - **Spring** (March, April, May)
  - **Summer** (June, July, August)
  - **Fall** (September, October, November)
  - This categorization will allow us to analyze seasonal trends in electricity usage and cost.

- **`is_holiday_season`**: Marks whether the month falls in the holiday season (November and December).  
  - **True** for November and December, **False** otherwise.  
  - This feature can help us assess the impact of holiday periods on electricity consumption.

- **`is_summer_peak`**: Indicates peak summer months, typically associated with high electricity usage due to cooling needs.
  - **True** for July and August, **False** otherwise.

- **`is_winter_peak`**: Identifies peak winter months, where heating needs might drive higher electricity consumption.
  - **True** for January and February, **False** otherwise.

By adding these features, we aim to capture seasonal, holiday, and peak period effects on electricity demand and pricing, which are essential for further analysis and modeling.


````python
import pandas as pd

# Define a function to categorize months into seasons based on traditional groupings
# Winter: December, January, February
# Spring: March, April, May
# Summer: June, July, August
# Fall: September, October, November
def assign_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Apply the season assignment function to the 'month' column of the DataFrame
# This will create a new 'season' column in final_data with values 'Winter', 'Spring', 'Summer', or 'Fall'
final_data['season'] = final_data['month'].apply(assign_season)

# Create additional columns indicating specific peak or holiday seasons
# Each column will contain True if the month is part of that season, and False otherwise

# Holiday season indicator: sets True for November and December, otherwise False
final_data['is_holiday_season'] = final_data['month'].apply(lambda x: True if x in [11, 12] else False)

# Summer peak indicator: sets True for July and August (common peak usage months), otherwise False
final_data['is_summer_peak'] = final_data['month'].apply(lambda x: True if x in [7, 8] else False)

# Winter peak indicator: sets True for January and February (common winter peak months), otherwise False
final_data['is_winter_peak'] = final_data['month'].apply(lambda x: True if x in [1, 2] else False)

# Display the updated DataFrame to confirm the changes and new columns
print("Updated DataFrame with time-based features as True/False indicators:")
print(final_data.head())
````

#### 2.5.2 Weather-Based Feature Engineering

In this section, we add new features based on weather characteristics to capture the impact of temperature, precipitation, and snowfall on electricity usage. These binary indicators help us analyze the effects of specific weather conditions on electricity demand and cost.

- **`is_high_temp`**: Marks hot weather days where the average temperature exceeds 80°F.
  - **True** if `avg_temp` > 80°F, **False** otherwise.

- **`is_low_temp`**: Identifies freezing conditions where the average temperature falls below 32°F.
  - **True** if `avg_temp` < 32°F, **False** otherwise.

- **`has_precipitation`**: Indicates any measurable precipitation.
  - **True** if `precipitation` > 0, **False** otherwise.

- **`is_high_precipitation`**: Marks heavy rainfall days with precipitation over 1 inch.
  - **True** if `precipitation` > 1 inch, **False** otherwise.

- **`has_snowfall`**: Identifies days with measurable snowfall.
  - **True** if `snowfall` > 0, **False** otherwise.

- **`is_heavy_snowfall`**: Flags significant snow accumulation with snowfall over 5 inches.
  - **True** if `snowfall` > 5 inches, **False** otherwise.

These weather-based features provide a more nuanced view of how different weather events influence electricity usage patterns. By capturing extreme temperatures, precipitation, and snowfall levels, we gain insights into how environmental factors drive demand for electricity.


````python
# Define thresholds based on weather characteristics to create binary weather indicators

# High temperature indicator: True if avg_temp exceeds a threshold (e.g., 80°F), indicating hot weather
final_data['is_high_temp'] = final_data['avg_temp'].apply(lambda x: True if x > 80 else False)

# Low temperature indicator: True if avg_temp is below a threshold (e.g., 32°F), indicating freezing conditions
final_data['is_low_temp'] = final_data['avg_temp'].apply(lambda x: True if x < 32 else False)

# Precipitation indicator: True if there was measurable precipitation, otherwise False
final_data['has_precipitation'] = final_data['precipitation'].apply(lambda x: True if x > 0 else False)

# High precipitation indicator: True if precipitation exceeds a threshold (e.g., 1 inch), indicating heavy rainfall
final_data['is_high_precipitation'] = final_data['precipitation'].apply(lambda x: True if x > 1 else False)

# Snowfall indicator: True if there was measurable snowfall, otherwise False
final_data['has_snowfall'] = final_data['snowfall'].apply(lambda x: True if x > 0 else False)

# Heavy snowfall indicator: True if snowfall exceeds a threshold (e.g., 5 inches), indicating significant snow accumulation
final_data['is_heavy_snowfall'] = final_data['snowfall'].apply(lambda x: True if x > 5 else False)

# Display the updated DataFrame to confirm the new weather-based features
print("Updated DataFrame with weather-based features as True/False indicators:")
print(final_data.head())
````

#### 2.5.3 Usage and Cost-Based Feature Engineering

In this section, we add features related to electricity usage, revenue, and price to better understand economic factors that influence electricity consumption. These metrics allow for a deeper analysis of customer behavior and the financial aspects of electricity consumption.

- **`usage_per_customer`**: Calculates the average electricity usage per customer by dividing total `sales` by the number of `customers`.
  - Helps in understanding individual customer consumption patterns.

- **`revenue_per_customer`**: Computes the average revenue generated per customer by dividing `revenue` by `customers`.
  - Provides insight into the financial contribution of each customer.

- **`usage_yoy_growth`**: Measures the year-over-year growth rate in electricity usage by calculating the percentage change in `sales` compared to the same month in the previous year.
  - This feature highlights usage trends and growth over time.

- **`price_yoy_growth`**: Tracks the year-over-year growth rate in electricity price by computing the percentage change in `price` compared to the same month in the previous year.
  - Useful for monitoring pricing trends and understanding how price changes impact customer usage.

These features enhance the dataset with information on customer usage patterns, revenue contributions, and growth trends, supporting a comprehensive analysis of the economic factors affecting electricity demand.


````python
import pandas as pd

# Usage and Cost-Based Feature Engineering

# 1. Calculate average usage per customer (sales divided by customers, handling division by zero)
final_data['usage_per_customer'] = final_data.apply(
    lambda row: row['sales'] / row['customers'] if row['customers'] != 0 else 0, axis=1)

# 2. Calculate average revenue per customer (revenue divided by customers, handling division by zero)
final_data['revenue_per_customer'] = final_data.apply(
    lambda row: row['revenue'] / row['customers'] if row['customers'] != 0 else 0, axis=1)

# 3. Calculate year-over-year growth rate of usage (change in sales compared to same month last year)
final_data['usage_yoy_growth'] = final_data.groupby('state')['sales'].pct_change(periods=12) * 100

# 4. Calculate year-over-year growth rate of price (change in price compared to same month last year)
final_data['price_yoy_growth'] = final_data.groupby('state')['price'].pct_change(periods=12) * 100

# Display the updated DataFrame with new features
print("Updated DataFrame with usage and cost-based features:")
print(final_data.head())
````

#### 3.2.1 Electricity Usage and Price Over Time

Our first analysis in trend analysis focuses on electricity usage across the United States over time. We will examine:
- **Annual Usage Trend**: Aggregating sales by year to observe the overall trend in electricity usage across the country.
- **Monthly Average Sales Over Years**: Calculating the average monthly sales for each year to detect any seasonal variations.

This analysis will allow us to visualize overall demand patterns, helping to identify periods of growth or decline in electricity consumption across the nation.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set style for better aesthetics
sns.set_style("whitegrid")

# Filter out 2024 data as it might be incomplete
filtered_data = final_data[final_data['year'] < 2024]

# Group by year and month to calculate the monthly average price and usage across all states
monthly_avg = filtered_data.groupby(['year', 'month']).agg({
    'price': 'mean',
    'sales': 'mean'
}).reset_index()

# Set up the figure
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot monthly average usage (sales) on the left y-axis
sns.lineplot(data=monthly_avg, x='year', y='sales', marker='o', color='b', ax=ax1, label="Average Monthly Sales (Usage)")
ax1.set_xlabel("Year")
ax1.set_ylabel("Average Monthly Sales (MWh)", color='b')  # in million kilowatt-hours
ax1.tick_params(axis='y', labelcolor='b')

# Plot monthly average price on the right y-axis
ax2 = ax1.twinx()
sns.lineplot(data=monthly_avg, x='year', y='price', marker='o', color='g', ax=ax2, label="Average Monthly Price")
ax2.set_ylabel("Average Monthly Price (cents/kWh)", color='g')  # in cents per kilowatt-hour
ax2.tick_params(axis='y', labelcolor='g')

# Add title and adjust legend positioning
plt.title("Monthly Average Electricity Usage and Price (2001-2023)")
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9), frameon=False)

# Show plot
plt.show()
````

**Analysis of Monthly Average Electricity Usage and Price (2001-2023)**

The graph above shows the trends in **average monthly electricity usage (in MWh)** and **average monthly price (in $)** over the years from 2001 to 2023.

- **Usage Trend (Blue Line)**:
  - The average monthly electricity usage shows a general upward trend from 2001 to around 2008, suggesting a growing demand for electricity.
  - Between 2008 and 2015, the trend appears to stabilize, with usage fluctuating slightly but remaining relatively steady.
  - From 2015 onward, we observe a slight decline, followed by a plateau, indicating possible saturation in demand or improved energy efficiency measures.

- **Price Trend (Green Line)**:
  - The average monthly price of electricity has steadily increased throughout the period, with a noticeable rise beginning around 2010.
  - Prices appear to increase more sharply after 2018, possibly due to inflationary pressures, policy changes, or shifts in energy production costs.
  - The price trend continues to increase into 2023, reflecting ongoing economic or supply chain impacts on electricity costs.

- **Relationship Between Usage and Price**:
  - There is an inverse relationship observed in some parts of the graph, particularly post-2015, where usage slightly declines or stabilizes while prices continue to rise. This could indicate that higher prices may be leading to reduced consumption or greater efficiency measures by consumers.
  - The shaded areas around each line represent the variability or uncertainty within each trend, showing that usage data has more variability over the years compared to price.

This analysis helps highlight key trends in electricity usage and cost, setting the stage for a deeper investigation into the factors driving these patterns, such as seasonality, state-wise variations, and economic factors.

#### 3.2.2 Customer Growth and Usage Per Customer
- **Total Customers Over Time:** Analyze the trend in the total number of customers over the years to see if there has been significant growth in electricity consumers.
- **Average Usage Per Customer Over Time:** Track usage_per_customer (sales per customer) over time. This metric could reveal whether individual consumption is increasing or decreasing, possibly due to energy efficiency measures or shifts in lifestyle.

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for better aesthetics
sns.set_style(style="whitegrid")

# Filter out data from 2001 to 2007 (missing customers data) and exclude 2024 for completeness
filtered_data = final_data[(final_data['year'] > 2007) & (final_data['year'] < 2024)]

# Group data by year to calculate total customers and average usage per customer
yearly_data = filtered_data.groupby('year').agg({
    'customers': 'sum',             # Sum of customers per year
    'usage_per_customer': 'mean'     # Average usage per customer per year
}).reset_index()

# Set up the figure and primary y-axis for total customers
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot total customers on the left y-axis
sns.lineplot(data=yearly_data, x='year', y='customers', marker='o', color='b', ax=ax1, label='Total Customers')
ax1.set_xlabel("Year")
ax1.set_ylabel("Total Customers", color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Create a secondary y-axis for average usage per customer
ax2 = ax1.twinx()
sns.lineplot(data=yearly_data, x='year', y='usage_per_customer', marker='o', color='g', ax=ax2, label='Average Usage per Customer')
ax2.set_ylabel("Average Usage per Customer (MWh)", color='g')
ax2.tick_params(axis='y', labelcolor='g')

# Set the title and adjust layout to prevent overlap
plt.title("Customer Growth and Average Usage per Customer Over Time (2008-2023)")
fig.tight_layout()

# Position legends for both y-axes
ax1.get_legend().set_bbox_to_anchor((0.15, 0.85))
ax2.get_legend().set_bbox_to_anchor((0.85, 0.85))

# Show the plot
plt.show()
````

**Customer Growth and Average Usage per Customer Analysis (2008-2023)**

In this analysis, we explored trends in the total number of customers and the average usage per customer over time. The data from 2008 onward was used because earlier years (2001-2007) had missing values in the `customers` column, which could skew the trends if included.

- **Total Customers**: The blue line shows a steady increase in the total number of electricity customers over the years. This growth could reflect population growth, urbanization, and increased electrification across the USA.
- **Average Usage per Customer**: The green line shows fluctuations in average electricity usage per customer, with notable declines after 2016 and slight recovery around 2020. This decrease in per-customer usage may be attributed to energy efficiency improvements, behavioral changes, or possibly economic factors impacting electricity consumption.

The chart illustrates a clear upward trend in customer numbers, while the per-customer usage trend remains more variable, suggesting that while more people are using electricity, individual consumption patterns may be shifting.

This can provide valuable context for understanding energy demands, cost variations, and weather influences over time.

Through this analysis, we aim to build a comprehensive understanding of how seasonal factors influence electricity consumption, pricing, and other related metrics.


##### 3.3.1 Seasonal Electricity Usage and Price 

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set style for better aesthetics
sns.set(style="whitegrid")

# Filter out data from 2024 for consistency if it's incomplete
filtered_data = final_data[final_data['year'] < 2024]

# Function to plot analysis based on binary seasonal indicators
def plot_seasonal_analysis(column_name, title):
    # Group data based on the column (True/False) to calculate average price and usage
    seasonal_avg = filtered_data.groupby(column_name).agg({
        'price': 'mean',
        'sales': 'mean'
    }).reset_index()

    # Set up the figure with dual y-axes
    fig, ax1 = plt.subplots(figsize=(10, 6))

    # Plot average sales (usage) on the left y-axis
    sns.barplot(data=seasonal_avg, x=column_name, y='sales', color='skyblue', ax=ax1)
    ax1.set_xlabel(f"{title} (True = Peak, False = Off-Peak)")
    ax1.set_ylabel("Average Sales (MWh)", color='b')
    ax1.tick_params(axis='y', labelcolor='b')

    # Create a second y-axis to plot average price
    ax2 = ax1.twinx()
    sns.lineplot(data=seasonal_avg, x=column_name, y='price', color='orange', marker='o', ax=ax2)
    ax2.set_ylabel("Average Price (cents/kWh)", color='orange')
    ax2.tick_params(axis='y', labelcolor='orange')

    # Title and show plot
    plt.title(f"Electricity Usage and Price during {title}")
    plt.show()

# Analyze each seasonal column
plot_seasonal_analysis('season', "seasons")
plot_seasonal_analysis('is_summer_peak', "Summer Peak Period")
plot_seasonal_analysis('is_winter_peak', "Winter Peak Period")
plot_seasonal_analysis('is_holiday_season', "Holiday Season")
````

**Electricity Usage and Price Analysis Based on Seasonal and Temporal Trends**

1. **Overall Seasonal Trends**  
   - **Usage**: Electricity usage is highest in the **Summer** season and slightly lower in **Winter** and **Fall**, with the lowest usage observed in **Spring**. This could be due to the increased need for air conditioning in summer and heating in winter.
   - **Price**: Electricity prices peak in **Summer**, possibly due to higher demand and the associated costs of meeting peak loads. Prices are at their lowest in **Winter** and **Spring**, suggesting less price pressure during these periods.

2. **Summer Peak Period**  
   - **Usage**: During the summer peak, average electricity usage is notably higher, reflecting increased air conditioning demand.
   - **Price**: Prices are also at their highest during the summer peak, aligning with the surge in demand, which may require additional resources or peak-hour electricity pricing mechanisms.

3. **Winter Peak Period**  
   - **Usage**: Electricity usage is consistently high during winter, even though it does not peak as drastically as in summer.
   - **Price**: Interestingly, prices are somewhat lower during the winter peak compared to summer. This may be due to a more stable supply or fewer cost pressures in the colder months.

4. **Holiday Season**  
   - **Usage**: The holiday season, spanning late November through December, sees steady electricity usage due to holiday gatherings, lighting, and heating needs.
   - **Price**: Prices dip slightly during the holiday season, possibly due to seasonal pricing adjustments or promotions to encourage consumption. The slight decrease in price may indicate a competitive adjustment rather than a true reflection of usage trends.

### 4.1 Model Selection and Evaluation

In our analysis, we opted for XGBoost as our primary machine learning model for predicting electricity usage. This decision was based on several factors that highlight its advantages over traditional models like linear regression, particularly in handling complex datasets with non-linear relationships.

**What is XGBoost?**

XGBoost, which stands for eXtreme Gradient Boosting, is an advanced implementation of the gradient boosting framework. To understand its significance, it's helpful to break down a few key concepts:

**Gradient Descent:** At its core, XGBoost builds on the concept of gradient descent, a popular optimization algorithm used in machine learning. Gradient descent aims to minimize the error of a model by iteratively adjusting its parameters based on the gradient (slope) of the loss function. This helps the model learn from its mistakes and improve accuracy over time.

**Boosting:** Boosting is a technique that combines multiple weak models (models that perform slightly better than random chance) to create a strong predictive model. In boosting, models are trained sequentially, with each new model focusing on the errors made by the previous ones. This iterative process enhances the overall performance of the ensemble.

**Ensemble Learning:** Ensemble learning refers to the method of combining multiple models to produce a more accurate prediction than any single model could achieve alone. XGBoost leverages this idea, using a collection of decision trees (the individual models) that work together to provide robust predictions.

**Why Choose XGBoost?**

**Handling Non-Linearity:** One of the main advantages of XGBoost over linear regression is its ability to capture non-linear relationships between features and the target variable. While linear regression assumes a straight-line relationship, XGBoost can model complex interactions and patterns in the data, making it particularly effective for real-world datasets like electricity usage.

**Robustness to Overfitting:** XGBoost incorporates regularization techniques (L1 and L2) that help prevent overfitting, which occurs when a model learns the training data too well and fails to generalize to new data. This robustness is crucial for making accurate predictions in practical applications.

**Speed and Efficiency:** XGBoost is designed for performance. It utilizes parallel processing and cache optimization, making it faster and more efficient than many other algorithms, including traditional boosting methods. This efficiency allows for quicker training times, which is particularly beneficial when working with large datasets.

**Feature Importance:** XGBoost provides insights into the importance of each feature in making predictions. This interpretability helps us understand which factors significantly influence electricity usage, aiding in decision-making and further analysis.

**Hyperparameter Tuning**: XGBoost offers a wide array of hyperparameters that can be fine-tuned to optimize model performance. This flexibility allows us to tailor the model to our specific dataset, enhancing prediction accuracy.

In summary, we chose XGBoost over other models like linear regression due to its ability to handle complex relationships, robustness against overfitting, superior speed and efficiency, interpretability of feature importance, and flexibility in hyperparameter tuning. These factors make XGBoost a powerful tool for our electricity usage predictions, aligning well with the challenges presented by our dataset.

#### 4.2.2 Choosing our Featues and Preparing the data

A. Primary Features (no missing values):

````python
primary_features = [
    'price', 'avg_temp', 'precipitation', 'total_population',
    'season', 'is_summer_peak', 'is_winter_peak',
    'is_high_temp', 'is_low_temp', 'has_precipitation',
    'is_high_precipitation', 'has_snowfall', 'is_heavy_snowfall'
]
````

B. Target Variable:

Let's use `sales` as our target (no missing values). Target Variable is what we aim to predict. This is extracted from the DataFrame and will be used in training the model.

We'll avoid usage_per_customer due to missing data

````python
# 1. Basic data preparation
def prepare_data(df):
    # Convert season to numeric using one-hot encoding
    season_dummies = pd.get_dummies(df['season'], prefix='season', dtype=int)
    
    # Combine features
    features = df[primary_features].copy()
    features = pd.concat([features, season_dummies], axis=1)
    features.drop('season', axis=1, inplace=True)
    
    # Target variable
    target = df['sales']
    
    return features, target

# 2. Time-based split
def create_time_splits(df, features, target):
    # Training data: 2008-2019
    train_mask = (df['year'] >= 2008) & (df['year'] <= 2019)
    # Validation data: 2020-2021
    val_mask = (df['year'] >= 2020) & (df['year'] <= 2021)
    # Test data: 2022-2023
    test_mask = (df['year'] >= 2022) & (df['year'] <= 2023)
    
    X_train = features[train_mask]
    y_train = target[train_mask]
    X_val = features[val_mask]
    y_val = target[val_mask]
    X_test = features[test_mask]
    y_test = target[test_mask]
    
    return X_train, X_val, X_test, y_train, y_val, y_test
````

Method `prepare_data`: This method prepares the data for our XGBoost model by first seasons into **one-hot encoding** which is basically just coverting each season into values of 0 and 1 so it is better for our model to understand it. So now we will have `is_summer` with a vlaue of either 0 (False) and 1 (True) rather than `season` = "Summer" (String).
The method also combines all the features into a dataframe and makes `sales` our target variable. Returns features as a dataframe and target as a column.

Method `create_time_splits`: This method is used for splitting our dataset into three different dataframes, most of our dataset will go into `training` the mode from 2008-2019 since for these years we have no missing values and our model will learn the best on it. We will use 2020 and 2021 for `validation`, which helps us fine-tune the model. This means we check how well the model performs on this data and make adjustments to improve it, like changing settings. It's important to keep the test data separate until the very end, so we only use it for the final check. Finally, we will `test` the model using the data from 2022 and 2023 to see how accurately it predicts actual results.

````python
# Prepare features and target
features, target = prepare_data(df)

# Split data
X_train, X_val, X_test, y_train, y_val, y_test = create_time_splits(df, features, target)

# Check our split sizes
print("Training set shape:", X_train.shape)
print("Validation set shape:", X_val.shape)
print("Test set shape:", X_test.shape)
y_test
````

#### 4.2.3 Creating our model

We used XGBRegressor because our goal is to predict continuous values (like electricity usage or sales amounts), making it the ideal model for regression tasks where the output is a number rather than a category or label.

````python
def create_final_model():
    model = xgb.XGBRegressor(
        n_estimators=200,          # More trees
        learning_rate=0.03,        # Even slower learning
        max_depth=5,               # Slightly deeper trees
        min_child_weight=3,        # Adjusted weight
        subsample=0.9,             # Use 90% of data
        colsample_bytree=0.9,      # Use 90% of features
        reg_alpha=0.1,             # L1 regularization
        reg_lambda=1,              # L2 regularization
        random_state=42,           # 42 is just a convention number this just tells data to be random consistently
        eval_metric='rmse'         # Root Mean Squared Error as evaluation metric
    )
    return model

# Create and train final model
model_final = create_final_model()
model_final.fit(
    X_train,
    y_train,
    eval_set=[(X_train, y_train), (X_val, y_val)],
    verbose=True
)
````

Method: `create_final_model`: The `create_final_model` function constructs an optimized XGBoost model with carefully chosen hyperparameters (I played around with the parameters a lot and chose the ones which gave the least error). This function returns a model instance configured for better performance through the following settings:

`n_estimators = 200:` The number of trees in the ensemble, which helps the model learn more patterns from the data.

`learning_rate = 0.03:` A slower learning rate to ensure the model learns more gradually, which helps in avoiding overfitting.

`max_depth = 5:` Slightly deeper trees to capture more complex patterns in the data while maintaining a balance to prevent overfitting.

`min_child_weight = 3:` Ensures that the model only splits nodes with sufficient weight to avoid overfitting.

`subsample=0.9:` Uses 90% of the training data for each tree, adding randomness to help generalize the model.

`colsample_bytree = 0.9:` Uses 90% of the features for each tree, adding diversity to the trees in the ensemble.

`reg_alpha=0.1:` L1 regularization term to help reduce overfitting by adding a penalty for complexity.

`reg_lambda=1:` L2 regularization term to further stabilize the model and reduce overfitting.

`random_state=42:` Ensures reproducibility of results.

`eval_metric='rmse':` Specifies the root mean squared error as the evaluation metric for better tracking of prediction accuracy.

#### 4.2.5 Using a Different Approach: **"Log Transform"**

Think of it like this:

**Regular Numbers vs. Log Transformed:**
- 1,000 → ~7
- 10,000 → ~9
- 100,000 → ~11

**Why Use Log Transform?**

**Makes Big Differences Smaller:**

*Regular difference:*
- 100,000 - 1,000 = 99,000 (huge gap!)

*Log difference:*
- 11 - 7 = 4 (much smaller gap!)

**Real Example from Our Data:**

*Before Log Transform:*
- Small state: 533 units
- Big state: 12,236 units
*(Very hard for the model to handle this big range)*

*After Log Transform:*
- Small state: ~6.3
- Big state: ~9.4
*(Easier for the model to handle this smaller range)*

**Think of It Like Converting Heights:**
- Regular: 48 inches vs. 72 inches (24-inch gap)
- Feet: 4 feet vs. 6 feet (2-foot gap)
- Same information, just smaller numbers!

**Benefits of Log Transform:**
- Reduces the scale difference between small and large values
- Makes the target distribution more normal
- Helps the model handle large values more effectively




````python
# Transform target variable
y_train_log = np.log1p(y_train)  # log1p preserves zero values
y_val_log = np.log1p(y_val)
y_test_log = np.log1p(y_test)

print(y_train.head(), y_train_log.head())

def create_log_model():
    model = xgb.XGBRegressor(
        n_estimators=200,
        learning_rate=0.03,
        max_depth=5,
        min_child_weight=3,
        subsample=0.9,
        colsample_bytree=0.9,
        reg_alpha=0.1,
        reg_lambda=1,
        random_state=42,
        eval_metric='rmse'
    )
    return model

# Train model on log-transformed data
model_log = create_log_model()
model_log.fit(
    X_train,
    y_train_log,
    eval_set=[(X_train, y_train_log), (X_val, y_val_log)],
    verbose=True
)

# Make predictions and transform back
train_pred_log = np.expm1(model_log.predict(X_train))
val_pred_log = np.expm1(model_log.predict(X_val))
test_pred_log = np.expm1(model_log.predict(X_test))

print(train_pred_log)
````

#### 4.2.6 A Newer and Better Approach: **"Stratified Approach"** - **Population-Based Strategy**

Our previous results highlighted that `total_population` plays a significant role in our predictions. To improve our model's performance, we can adopt a population-based approach. The idea is simple: the United States has states with vastly different population sizes. More population often translates to higher electricity usage, so using a single model for all states may not be ideal.

**Why This Approach?**

Splitting the states based on population size allows each model to specialize and improve prediction accuracy within its category. We can classify states into three distinct groups:
- `Small` states (like Wyoming)
- `Medium` states (like Arizona)
- `Large` states (like California)

**Benefits of This Approach:**
- Each model becomes an "expert" for its specific group size.
- Helps tailor predictions to better fit the characteristics of each category.

**Think of It Like This:**
- A small state specialist handles low-population areas.
- A medium state specialist focuses on states with moderate populations.
- A large state specialist tackles predictions for highly populated states.

By segmenting the data this way, we can create more targeted models that better capture the unique patterns within each group, leading to improved overall performance.


#### First let's start by splitting states into `small`, `medium` and `large`

````python
# 1. First, create population groups
X_train['population_group'] = pd.qcut(X_train['total_population'], 
                                    q=3, 
                                    labels=['small', 'medium', 'large'])

# 2. Let's see how many states in each group
print("States in each population group:")
print(X_train['population_group'].value_counts())

# 3. Let's see average sales for each group
print("\nAverage sales by population group:")
for group in ['small', 'medium', 'large']:
    mask = X_train['population_group'] == group
    avg_sales = y_train[mask].mean()
    print(f"{group} population states: {avg_sales:.2f}")
````

#### Now let's code up our new model!

````python
# First, let's create a function to add population groups
def add_population_groups(data):
    data = data.copy()
    data['population_group'] = pd.qcut(data['total_population'], 
                                     q=3, 
                                     labels=['small', 'medium', 'large'])
    return data

# Add population groups to all our datasets
X_train = add_population_groups(X_train)
X_val = add_population_groups(X_val)
X_test = add_population_groups(X_test)



# Now let's create our models
def create_stratified_models():
    models = {}
    
    for group in ['small', 'medium', 'large']:
        print(f"\nTraining model for {group} population states...")
        
        # Get data for this group
        mask = X_train['population_group'] == group
        X_group = X_train[mask].drop(['population_group'], axis=1)
        y_group = y_train[mask]
        
        # Create and train model
        model = xgb.XGBRegressor(
            n_estimators=100,
            learning_rate=0.05,
            max_depth=4,
            min_child_weight=3,
            subsample=0.8,
            colsample_bytree=0.8,
            random_state=42,
            eval_metric='rmse'
        )
        
        # Train the model
        model.fit(X_group, y_group, verbose=False)
        models[group] = model
    
    return models

# Train our models
stratified_models = create_stratified_models()

# Make predictions for each group
print("\nModel Performance by Group:")
for group in ['small', 'medium', 'large']:
    # Training performance
    train_mask = X_train['population_group'] == group
    X_group_train = X_train[train_mask].drop(['population_group'], axis=1)
    train_pred = stratified_models[group].predict(X_group_train)
    train_rmse = root_mean_squared_error(y_train[train_mask], train_pred)
    
    # Test performance
    test_mask = X_test['population_group'] == group
    X_group_test = X_test[test_mask].drop(['population_group'], axis=1)
    test_pred = stratified_models[group].predict(X_group_test)
    test_rmse = root_mean_squared_error(y_test[test_mask], test_pred)
    
    print(f"\n{group.capitalize()} Population States:")
    print(f"Training RMSE: {train_rmse:.2f}")
    print(f"Test RMSE: {test_rmse:.2f}")
    
    # Show sample predictions
    print(f"\nSample predictions for {group} states:")
    sample_df = pd.DataFrame({
        'Actual': y_test[test_mask][:3],
        'Predicted': test_pred[:3],
        'Error%': abs(y_test[test_mask][:3] - test_pred[:3]) / y_test[test_mask][:3] * 100
    })
    print(sample_df)
````

#### 4.2.10 Create and Visualize Data for 2024 and 2025

In this section, we will predict electricity usage for the years **2024** and **2025** based on the model's performance with the actual vs predicted usage for previous years. The model has shown good accuracy, and now we will extend these predictions to future years.

The approach involves the following steps:

- **Data remains constant** for the most part, with the primary variables being **price** and **weather temperature**, which we observe change over time.
- To predict future values, we will analyze the changes in **price** and **temperature** between **January 2022** and **January 2023**. Using these changes as a reference, we will calculate the average changes for these features and apply the same adjustments for **2024** and **2025**.
- All other features, such as population and usage patterns, will remain constant, as these are less likely to change significantly year over year.
- Using these modified values, we will predict the electricity usage for **2024** and **2025**.

Finally, we will **visualize** the predictions by plotting the results for each state, similar to the approach for the previous years. These predictions will provide insights into potential future energy demands, helping with long-term planning and resource allocation.


##### Let's start by reviewing the values for each month of 2022 and 2023 for Massachusetts (as an example).

````python
# Create a combined DataFrame for analysis
combined_data = test_predictions.copy()
combined_data['price'] = X_test['price']
combined_data['avg_temp'] = X_test['avg_temp']
combined_data['precipitation'] = X_test['precipitation']

# Now let's analyze changes for one state (let's use Massachusetts as example)
print("Changes from 2022 to 2023 for Massachusetts:")
print("\nMonth by Month comparison:")
state_data = combined_data[combined_data['state'] == 'Massachusetts'].sort_values(['year', 'month'])

for month in range(1, 13):
    data_2022 = state_data[(state_data['year'] == 2022) & (state_data['month'] == month)]
    data_2023 = state_data[(state_data['year'] == 2023) & (state_data['month'] == month)]
    
    if not data_2022.empty and not data_2023.empty:
        print(f"\nMonth {month}:")
        print(f"Temperature: {data_2022['avg_temp'].iloc[0]:.2f}°F -> {data_2023['avg_temp'].iloc[0]:.2f}°F")
        print(f"Price: ${data_2022['price'].iloc[0]:.2f} -> ${data_2023['price'].iloc[0]:.2f}")
        print(f"Precipitation: {data_2022['precipitation'].iloc[0]:.2f} -> {data_2023['precipitation'].iloc[0]:.2f}")

# Calculate average changes
temp_change = (state_data[state_data['year'] == 2023]['avg_temp'].mean() - 
               state_data[state_data['year'] == 2022]['avg_temp'].mean())
price_change = (state_data[state_data['year'] == 2023]['price'].mean() - 
                state_data[state_data['year'] == 2022]['price'].mean())

print("\nAverage Changes:")
print(f"Temperature change: {temp_change:.2f}°F")
print(f"Price change: ${price_change:.2f}")
````

#### Analysis

We can see that there is a difference in our features from 2022 - 2023. Now let's take average for every single state from 2022 - 2023

````python
# Calculate state-specific average changes
state_changes = {}

for state in combined_data['state'].unique():
    state_data_2022 = combined_data[(combined_data['state'] == state) & (combined_data['year'] == 2022)]
    state_data_2023 = combined_data[(combined_data['state'] == state) & (combined_data['year'] == 2023)]
    
    # Calculate average changes for this state
    temp_change = (state_data_2023['avg_temp'].mean() - state_data_2022['avg_temp'].mean())
    price_change = (state_data_2023['price'].mean() - state_data_2022['price'].mean())
    
    state_changes[state] = {
        'temp_change': temp_change,
        'price_change': price_change
    }

# Let's look at how changes vary by state
print("Average changes by state:")
for state, changes in state_changes.items():
    print(f"\n{state}:")
    print(f"Temperature change: {changes['temp_change']:.2f}°F")
    print(f"Price change: ${changes['price_change']:.2f}")
````

#### Capping Extreme Changes for Realistic Projections

Upon reviewing the average changes in **temperature** and **price** for each state, some states, such as Delaware, showed unusually high jumps (e.g., a **5°F** increase). These significant changes are not realistic, as it's unlikely for temperature or price to increase so drastically each year.

To ensure that our future projections remain realistic, the following caps were applied:

- **Temperature changes** are limited to a maximum of **±2°F**.
- **Price changes** are limited to a maximum of **±$2**.

These caps prevent any unrealistic spikes in our predictions, thereby ensuring more reliable future forecasts.

**Adjustment Summary:**

- **Before Capping:**
  - States with temperature changes exceeding ±2°F were identified.
  - States with price changes exceeding ±$2 were flagged.
  
- **After Capping:**
  - Temperature changes are now constrained to **±2°F**.
  - Price changes are limited to **±$2**.

By capping these values, we aim to avoid anomalies in future data and maintain realistic projections.


````python
print("Before capping:")
for state, changes in state_changes.items():
    if abs(changes['temp_change']) > 2 or abs(changes['price_change']) > 2:
        print(f"{state}:")
        print(f"  Temperature change: {changes['temp_change']:.2f}°F")
        print(f"  Price change: ${changes['price_change']:.2f}")

# Apply caps
for state in state_changes:
    # Cap temperature changes at ±2°F
    if state_changes[state]['temp_change'] > 2:
        state_changes[state]['temp_change'] = 2
    elif state_changes[state]['temp_change'] < -2:
        state_changes[state]['temp_change'] = -2
        
    # Cap price changes at ±$2
    if state_changes[state]['price_change'] > 2:
        state_changes[state]['price_change'] = 2
    elif state_changes[state]['price_change'] < -2:
        state_changes[state]['price_change'] = -2

print("\nAfter capping:")
for state, changes in state_changes.items():
    if abs(changes['temp_change']) == 2 or abs(changes['price_change']) == 2:
        print(f"{state}:")
        print(f"  Temperature change: {changes['temp_change']:.2f}°F")
        print(f"  Price change: ${changes['price_change']:.2f}")
````

#### Creating Future Datasets for 2024 and 2025

With our adjustments in place, we now aim to generate future datasets for **2024** and **2025**. The goal is to mirror the structure of our original dataset while incorporating projected changes in **temperature** and **price** based on our previous analysis.

To achieve this, we will:

- **Use 2023 data as a base** and apply projected changes for each state.
- Adjust **temperature** and **price** based on the average changes we calculated earlier, while ensuring these changes do not exceed our capped limits.
- Introduce **seasonal and weather-related features** to match the original dataset structure.

##### Steps Involved:

1. **Copy Base Data:**
   - We start by copying the dataset for **2023** and set the year to **2024** or **2025** accordingly.

2. **Apply State-Specific Changes:**
   - For each state, adjust the **average temperature** and **price** using the capped changes:
     - For **2024**, apply the changes once.
     - For **2025**, apply the changes cumulatively for two years.

3. **Add Seasonal Features:**
   - Introduce new columns to capture seasonal and weather-related information:
     - **Summer and Winter Peaks**: Identify summer (June, July, August) and winter (December, January, February) peak months.
     - **Seasons**: Categorize each month into Spring, Summer, Fall, or Winter.
     - **Temperature Extremes**: Mark months with temperatures significantly above or below the average.
     - **Precipitation and Snowfall**: Indicate whether a month typically has precipitation or heavy snowfall.

````python
base_2023 = combined_data[combined_data['year'] == 2023].copy()

def create_future_features(year, base_year_data, year_number, state_changes):
    """
    Create future features with capped state-specific changes
    """
    future_data = base_year_data.copy()
    future_data['year'] = year
    
    # Apply state-specific changes
    for state in future_data['state'].unique():
        state_mask = future_data['state'] == state
        changes = state_changes[state]
        
        # Apply cumulative changes for each state
        future_data.loc[state_mask, 'avg_temp'] = (
            future_data.loc[state_mask, 'avg_temp'] + 
            (changes['temp_change'] * year_number)
        )
        future_data.loc[state_mask, 'price'] = (
            future_data.loc[state_mask, 'price'] + 
            (changes['price_change'] * year_number)
        )
        
    # Add seasonal features
    for month in range(1, 13):
        month_mask = future_data['month'] == month
        
        # Summer/Winter peak
        future_data.loc[month_mask, 'is_summer_peak'] = month in [6, 7, 8]
        future_data.loc[month_mask, 'is_winter_peak'] = month in [12, 1, 2]
        
        # Seasons
        future_data.loc[month_mask, 'season_Spring'] = month in [3, 4, 5]
        future_data.loc[month_mask, 'season_Summer'] = month in [6, 7, 8]
        future_data.loc[month_mask, 'season_Fall'] = month in [9, 10, 11]
        future_data.loc[month_mask, 'season_Winter'] = month in [12, 1, 2]
        
        # Temperature features
        temps = future_data.loc[month_mask, 'avg_temp']
        future_data.loc[month_mask, 'is_high_temp'] = temps > temps.mean() + temps.std()
        future_data.loc[month_mask, 'is_low_temp'] = temps < temps.mean() - temps.std()
        
        # Precipitation features
        future_data.loc[month_mask, 'has_precipitation'] = True
        future_data.loc[month_mask, 'is_high_precipitation'] = True
        future_data.loc[month_mask, 'has_snowfall'] = month in [12, 1, 2, 3]
        future_data.loc[month_mask, 'is_heavy_snowfall'] = month in [12, 1, 2]
    
    return future_data

# Create 2024 and 2025 data
future_2024_capped = create_future_features(2024, base_2023, 1, state_changes)
future_2025_capped = create_future_features(2025, base_2023, 2, state_changes)
````

#### Predicting Electricity Usage for 2024 and 2025

With our future datasets for **2024** and **2025** prepared, the next step is to predict electricity usage using the **stratified models** based on population groups. Each state belongs to a specific population group, and we have trained separate models for each group. 

##### Approach:
- Use the stratified models to make predictions for each state's electricity usage in 2024 and 2025.
- The prediction process involves:
  - Iterating through each row in the dataset.
  - Determining the **population group** for the state.
  - Using the corresponding model to predict electricity usage based on the state's features.

````python
def make_predictions_for_future(future_data, stratified_models):
    predictions = []
    
    for idx, row in future_data.iterrows():
        group = row['population_group']
        features = row[X_test.columns].drop('population_group').values.reshape(1, -1)
        pred = stratified_models[group].predict(features)[0]
        predictions.append(pred)
    
    return predictions

# Add predictions
future_2024_capped['predicted_usage'] = make_predictions_for_future(future_2024_capped, stratified_models)
future_2025_capped['predicted_usage'] = make_predictions_for_future(future_2025_capped, stratified_models)
````
